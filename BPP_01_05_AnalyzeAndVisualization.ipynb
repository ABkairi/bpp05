```python
import pandas as pd
import matplotlib.pyplot as plt
!wget https://raw.githubusercontent.com/Abdullahfoysal/pp11-Project-AnalyzeAndVisualize/refs/heads/main/student_data.csv
#Script for no. 1 question
df=pd.read_csv('student_data.csv')
highest_5=df.nlargest(5,'cgpa')
print('Deatils of The top 5 students with the highest CGPA:\n\n',highest_5.sample(1))

#Script for no. 2 question

mask=df['cgpa']>3.8
print('\nStudents who have a CGPA greater than 3.8:\n\n',df[mask].sample(1))

#Script for no.3 question

df_male = df[df['gender'] == 'Male']
average_gpa_male = df_male['hsc_gpa'].mean()
print("\nAverage HSC GPA of Male Students:\n", average_gpa_male)

#Script for no. 4 question
df_missing_value = df.fillna({'gender': 'other'})
print('\nReplacing missing values in the "gender" column with "Other:\n',df_missing_value.sample(1))

#Script for no. question 5
start_year=2013
end_year = 2024

df_filtered = df[(df['passing_year'] >= start_year) & (df['passing_year'] <= end_year)]

graduation_counts = df_filtered['passing_year'].value_counts().sort_index()

plt.figure(figsize=(10, 6))
plt.plot(graduation_counts.index, graduation_counts.values, marker='o', linestyle='-', color='b')


plt.title('Number of Graduations (2013 to 2023)')
plt.xlabel('Year')
plt.ylabel('Number of Graduations')
plt.grid(True)
plt.xticks(graduation_counts.index)


plt.show()




!wget https://raw.githubusercontent.com/Abdullahfoysal/pp11-Project-AnalyzeAndVisualize/refs/heads/main/student_data.csv
```
